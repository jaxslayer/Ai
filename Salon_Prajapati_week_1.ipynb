```python
import numpy as np

# Given matrices
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# 1. Prove A * A^-1 = I (Identity matrix)
# Compute A^-1 (inverse of A)
A_inv = np.linalg.inv(A)
identity_matrix = np.dot(A, A_inv)  # Compute A * A^-1
print("1. A * A^-1:\n", identity_matrix)
print("Is A * A^-1 equal to I?:", np.allclose(identity_matrix, np.eye(2)))

# 2. Prove AB ≠ BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\n2. AB:\n", AB)
print("BA:\n", BA)
print("Is AB equal to BA?:", np.array_equal(AB, BA))

# 3. Prove (AB)ᵀ = BᵀAᵀ
AB_T = np.transpose(np.dot(A, B))  # Compute transpose of AB
BT_AT = np.dot(np.transpose(B), np.transpose(A))  # Compute Bᵀ * Aᵀ
print("\n3. (AB)ᵀ:\n", AB_T)
print("BᵀAᵀ:\n", BT_AT)
print("Is (AB)ᵀ equal to BᵀAᵀ?:", np.allclose(AB_T, BT_AT))
```

```python
#For the following arrays:
#A = np.array([[3,4],[7,8]]) and B = np.array([[5,3],[2,1]]);
#Prove following with Numpy:
A = np.array([[2, -3, 1],
              [1, -1, 2],
              [3, 1, -1]])

B = np.array([-1, -3, 9])

#Compute the determinant of A manually
det_A = (A[0, 0] * (A[1, 1] * A[2, 2] - A[1, 2] * A[2, 1]) -
         A[0, 1] * (A[1, 0] * A[2, 2] - A[1, 2] * A[2, 0]) +
         A[0, 2] * (A[1, 0] * A[2, 1] - A[1, 1] * A[2, 0]))

print("Determinant of A:", det_A)

#Compute the cofactor matrix

cofactor_matrix = np.array([
    [(A[1, 1] * A[2, 2] - A[1, 2] * A[2, 1]), -(A[1, 0] * A[2, 2] - A[1, 2] * A[2, 0]), (A[1, 0] * A[2, 1] - A[1, 1] * A[2, 0])],
    [-(A[0, 1] * A[2, 2] - A[0, 2] * A[2, 1]), (A[0, 0] * A[2, 2] - A[0, 2] * A[2, 0]), -(A[0, 0] * A[2, 1] - A[0, 1] * A[2, 0])],
    [(A[0, 1] * A[1, 2] - A[0, 2] * A[1, 1]), -(A[0, 0] * A[1, 2] - A[0, 2] * A[1, 0]), (A[0, 0] * A[1, 1] - A[0, 1] * A[1, 0])]
])

#Compute the adjoint (transpose of cofactor matrix)
adjoint_A = cofactor_matrix.T

#Compute the inverse of A
inverse_A = adjoint_A / det_A
print("\nInverse of A:\n", inverse_A)

#Solve for X = A⁻¹B
X = np.dot(inverse_A, B)
print("\nSolution (x, y, z):", X)
```

```python
import numpy as np
import random
import time

# Define the matrix size
size = 100

# Generate two 1000x1000 matrices using Python lists
matrix_a = [[random.random() for _ in range(size)] for _ in range(size)]
matrix_b = [[random.random() for _ in range(size)] for _ in range(size)]

# Function to perform matrix multiplication using lists
def matrix_multiplication_lists(matrix_a, matrix_b):
    result = [[0 for _ in range(len(matrix_b[0]))] for _ in range(len(matrix_a))]
    for i in range(len(matrix_a)):
        for j in range(len(matrix_b[0])):
            for k in range(len(matrix_b)):
                result[i][j] += matrix_a[i][k] * matrix_b[k][j]
    return result

# Measure time for Python list-based multiplication
start_time = time.time()
matrix_multiplication_lists(matrix_a, matrix_b)
list_time = time.time() - start_time
print(f"Python Lists Time: {list_time:.2f} seconds")

# Generate two 1000x1000 matrices using NumPy
matrix_a_np = np.random.rand(size, size)
matrix_b_np = np.random.rand(size, size)

# Measure time for NumPy array-based multiplication
start_time = time.time()
np.dot(matrix_a_np, matrix_b_np)
numpy_time = time.time() - start_time
print(f"NumPy Time: {numpy_time:.2f} seconds")
```
